### Simple Workflow/Graph using langgraph

 State: serves an input schema for all noes and edges

```python
from typing_extensions import TypedDict
class State(TypedDict):
    graphInfo: str
```

Nodes: Python functions/task

The frist positional argument is the state

```python
# Root Node 

def start_play(state: State):
    print("Start node has been called")  #doing some task/computation
    return {"graphInfo": state["graphInfo"] + " I will play"}   #updating the state with new info
```

Middle Nodes

```python
def badminton(state: State):
    print("Badminton")
    return {"graphInfo": state["graphInfo"] + "badminton"}
```

```python
def football(state: State):
    print("Football")
    return {"graphInfo": state["graphInfo"] + "football"}
```

Condition for moving from root to an edge

```python
import random 
from typing import Literal

#some conditional logic to decide which sport to play
def decide_sport(state: State) -> Literal["badminton", "football"]:
    
    if random.random() > 0.5:
        return "badminton"
    else:
        return "football"
```

### Building the graph 
Nodes and conditional logic have been defined

The StateGraph class is the graph class that we can use.

First, we initialize a StateGraph with the State class we defined above.

Then, we add our nodes and edges.

We use the START Node, a special node that sends user input to the graph, to indicate where to start our graph.

The END Node is a special node that represents a terminal node.

Finally, we compile our graph to perform a few basic checks on the graph structure.
